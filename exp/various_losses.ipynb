```python
import torch
from torch import Tensor
from matplotlib import pyplot as plt
```

### Quantile Loss

```python
quantile = 0.99

def quantile_loss(pred: Tensor, truth: Tensor, q: float):
    error = pred - truth
    return torch.mean(torch.max(q * error, ((q - 1) * error))**1)
```

```python
predictions = torch.arange(0, 2, 0.05)
```

```python
truth = torch.ones_like(predictions) * 1.05
```

```python
q_losses = [quantile_loss(p,t,quantile).item() for p,t in zip(predictions, truth)]
```

```python
plt.plot(predictions.numpy(), q_losses)
```
