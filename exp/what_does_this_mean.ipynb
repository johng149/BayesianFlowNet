```python
import torch
```

```python
batch_sizes = [1*(10**i) for i in range(6)]
```

```python
maximums = []
minimums = []
```

```python
for b in batch_sizes:
    dist = torch.normal(0, 1, size=(b, 128, 37))
    print(dist.shape)
    mean = torch.mean(dist, dim=0)
    maximums.append(torch.max(torch.abs(mean)).item())
    minimums.append(torch.min(mean).item())
    print(f"Batch size: {b}, Max abs mean: {maximums[-1]}, Min abs mean: {minimums[-1]}")
```
