```python
import torch
from torch import Tensor
```

```python
torch.manual_seed(0)
```

```python
alpha = torch.tensor([0.5])
```

```python
target = torch.tensor([
    [0.0, 1.0, 0.0],
    [1.0, 0.0, 0.0]
])
```

```python
seq_len, K = target.shape
```

```python
model_output = torch.randn(seq_len, K)
```

```python
seq_norm_loss = torch.sum((target.unsqueeze(0) - model_output.unsqueeze(0)) ** 2, dim=(-2, -1)) * 0.5 * K * alpha / seq_len
```

```python
torch.mean(seq_norm_loss.view(-1, 1))
```

```python
target_batched = target.unsqueeze(0).repeat(2, 1, 1)
model_output_batched = model_output.unsqueeze(0).repeat(2, 1, 1)
alpha_batched = alpha.repeat(2)
```

```python
batch_seq_norm_loss = torch.sum((target_batched - model_output_batched) ** 2, dim=(-2, -1)) * 0.5 * K * alpha_batched / seq_len
```

```python
batch_seq_norm_loss
```

```python
torch.mean(batch_seq_norm_loss.view(-1, 1))
```
