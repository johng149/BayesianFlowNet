```python
import torch
from torch.nn import functional as F
```

```python
batch_size = 10
alphas = torch.rand(batch_size)
t = torch.rand(batch_size)
```

```python
alphas.shape
```

```python
alphas, t
```

```python
# alphas and t should have one-to-one relationship, we sort t and then sort alpha accordingly
t_v, t_i = torch.sort(t)
alphas = alphas[t_i]
```

```python
t_v, t_i, alphas
```

```python
F.relu(alphas[:-1] - alphas[1:])
```
