```python
import torch
from torch.nn import functional as F
```

```python
b,s,k = 2,4,5
```

```python
mask = torch.tensor([
    [True, False, True, False],
    [False, True, True, True]
])
```

```python
res = torch.rand(b, s, k)
targ = F.one_hot(torch.randint(0, k, (b, s)), num_classes=k)
```

```python
pred = res.argmax(dim=-1)
target = targ.argmax(dim=-1)
```

```python
pred.shape, target.shape
```

```python
(pred == target)[mask].float().mean().item()
```
